## CSV

```python
from io import StringIO
```

```python
data=('col1,col2,col3\n'
      'x,y,1\n'
      'a,b,2\n'
      'c,d,3,\n')
      
```

```python
type(data)
```

```python
StringIO(data)
```

```python
df=pd.read_csv(StringIO(data), usecols=lambda x: x.upper() in ['COL1', 'COL3'])
```

```python
df.to_csv('Test.csv')
```
