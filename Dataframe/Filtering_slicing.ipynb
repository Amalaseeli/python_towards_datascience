```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
```

## Filtering and Selecting

```python
numbers_df=DataFrame(np.arange(0,90,3).reshape(10,3),index=['row1','row2','row3','row4','row5','row6','row7','row8','row9','row10'],columns=["column1",'column2','column3'])
numbers_df
```

## Types of Indexing
1. Simple indexing
2. Boolean indexing
3. Fancy indexing

```python
numbers_df.iloc[0,1]
```

```python
numbers_df.iloc[0,1]=20
```

```python
numbers_df
```

## Fancy indexing

```python
numbers_df.iloc[[1,2,4],[1,2]]
```

### Boolean indexing
#### Passing comparision operators

```python
mask=numbers_df>30
mask
```

```python
numbers_df[mask]
```

```python
numbers_df[numbers_df>30]=0
numbers_df
```

### Slicing

```python
numbers_df.iloc[2:5, 1:]
```
